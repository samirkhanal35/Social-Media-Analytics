```python
with open("YouTubeTrailerComments.txt", 'r', encoding='latin-1') as file:
    lines = file.readlines()
    
```

```python
import re
```

```python
lines
```

```python
# Define the regex pattern
pattern = r'\d+ \d+ \d+ \d+ \d+\n'
```

```python
len(lines)
```

```python
Total_data = []
index = 1
for line in lines:
    each_data = []
    if re.search(pattern, line):
        datas = line.strip('\n').split(' ')
        each_data = [int(i) for i in datas]
        each_data.insert(0, index)
        Total_data.append(each_data)
        index += 1

Total_data
```
