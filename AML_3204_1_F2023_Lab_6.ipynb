```python
import pandas as pd
from pytrends.request import TrendReq
```

```python
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['Lionel Messi'])
```

```python
df = pytrend.interest_by_region()
df
```

```python
df_indexed = df.sort_index()
```

```python
df_sorted = df_indexed.sort_values("Lionel Messi", ascending=False)
df_sorted
```

```python
# Plotting the data with MultiIndex
import matplotlib.pyplot as plt
df_sorted.head(20).plot(figsize=(12, 6), kind='bar', legend=False)
plt.title('Interest in Lionel Messi by Country')
plt.xlabel('Country')
plt.ylabel('Interest Score')
plt.show()
```

```python
# Search Items
brands = ["BMW", "Mercedes", "Ford", "Toyota", "Honda"]
```

```python
# build payload
pytrend.build_payload(kw_list=brands,
                      timeframe='today 12-m',
                      geo='CA')
```

```python
df_ca = pytrend.interest_by_region()
df_ca
```

```python
df_ca.plot(figsize=(12,6))
```

```python
df_ca.plot(figsize=(12,6), kind='bar')
```
