```python
from googleapiclient.discovery import build 
import pandas as pd 
import numpy as np 
```

```python
# Create a Youtube resource object
myYoutube = build('youtube', 'v3', developerKey=api_key)
```

```python
videoId= 'C6sVGS5NJG8'

video_response = myYoutube.commentThreads().list(
                    part='snippet, replies',
                    maxResults=200,
                    videoId=videoId).execute()
```

```python
video_response
```

```python
df_arr = []
for each in video_response['items']:
    temp = []
    # print(each)
    temp.append(each['snippet']['topLevelComment']['snippet']['textDisplay'])
    df_arr.append(temp)
```

```python
df = pd.DataFrame(df_arr, columns=['comments'])
df 
```
