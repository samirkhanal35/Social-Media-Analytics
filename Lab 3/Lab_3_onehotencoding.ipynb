```python
import numpy as np 
import pandas as pd 
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("data.csv")
dataset.head()
```

```python
# dropping the categorical column
tmp_dataset = dataset.copy()
db = dataset.copy()
dataset = dataset.drop(['Relationship Status'], axis=1)
dataset.head()
```

```python
y = dataset['Income']
X = dataset.drop(['Income'], axis=1)
```

```python
y
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.svm import SVC
model = SVC().fit(X_train, y_train)
```

```python
print(model.score(X_train, y_train))
```

```python
y_pred = model.predict(X_test)
y_ptrain = model.predict(X_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

mae = mean_absolute_error(y_true=y_train,y_pred=y_ptrain)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_train,y_pred=y_ptrain) #default=True
rmse = mean_squared_error(y_true=y_train,y_pred=y_ptrain,squared=False)
print("Errors for train set") 
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)

mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
print("Errors for test set") 
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)

```

```python
tmp_dataset = tmp_dataset.drop(['Relationship Status'], axis=1)
tmp_dataset.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
y = tmp_dataset['Income']
X = tmp_dataset.drop(['Income'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
 
regr = LogisticRegression(random_state=0)
regr.fit(X_train, y_train)
```

```python
print("Accuracy score for train set:")
print(regr.score(X_train, y_train))

print("Accuracy score for test set:")
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)
y_ptrain = regr.predict(X_train)
```

```python
mae = mean_absolute_error(y_true=y_train,y_pred=y_ptrain)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_train,y_pred=y_ptrain) #default=True
rmse = mean_squared_error(y_true=y_train,y_pred=y_ptrain,squared=False)
print("Errors for train set") 
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)

mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
print("Errors for test set") 
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```
